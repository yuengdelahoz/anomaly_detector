# Anomaly:
### something that deviates from what is standard, normal, or expected.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf

from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, losses
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Model
```

### Load data

```python
# Download the dataset
dataframe = pd.read_csv('dummy_data.csv', header=None)
raw_data = dataframe.values
dataframe.head()
```

### Normalize the data to [0,1]

```python
min_vals = tf.reduce_min(raw_data, axis=0)
max_vals = tf.reduce_max(raw_data, axis=0)
norm_data = (raw_data - min_vals) / (max_vals - min_vals)
pd.DataFrame(norm_data).head()
```

## Autoencoder

```python
class AnomalyDetector(Model):
  def __init__(self):
    super(AnomalyDetector, self).__init__()
    self.encoder = tf.keras.Sequential([
        layers.Dense(6, activation="relu"),
        layers.Dense(4, activation="relu"),
        layers.Dense(2, activation="linear")]) 
    
    self.decoder = tf.keras.Sequential([
        layers.Dense(4, activation="relu"),
        layers.Dense(6, activation="relu"),
        layers.Dense(8, activation="sigmoid")])
    
  def call(self, x):
    encoded = self.encoder(x)
    decoded = self.decoder(encoded)
    return decoded

```

```python
autoencoder = AnomalyDetector()
callbacks = [tf.keras.callbacks.EarlyStopping(monitor="loss",patience=2,verbose=1)]
autoencoder.compile(optimizer='adam', loss='mse')
history = autoencoder.fit(norm_data, norm_data, 
                          epochs=1000, 
                          batch_size=512,
                          shuffle=True,
                          callbacks = callbacks)
plt.plot(history.history["loss"], label="Training Loss")
plt.legend()
```

# Anomaly Detection:
- Most of the samples in the dataset belong to "normal data". Anomalies are outliers in the dataset.
- When training the model on a dataset that consists of mainly normal data, the model learns to reconstruct only normal data samples with very little reconstruction error.
- Normal instances will have a low reconstruction loss, and abnormal (outliers) a high (higher than normal) reconstruction loss

```python
reconstructions = autoencoder.predict(norm_data)
reconstruction_losses = tf.keras.losses.mse(reconstructions, norm_data)
```

## Solution #1
We use a histogram to count the frequency of normal and abnormal data (outliers).
High scores corresponds to low frequency — the evidence of outliers. We select a threshold that represents the cutoff point between normal and outliers.


```python
plt.hist(reconstruction_losses[None,:], bins=5, rwidth=0.8)
plt.xlabel("Train loss")
plt.ylabel("No of examples")
plt.show()
```

```python
cut_off_threshold = 0.04
anomalies_found_1 = reconstruction_losses > cut_off_threshold
anomaly_indices_1 = np.squeeze(np.where(anomalies_found_1))
anomalies_1 = raw_data[anomaly_indices_1] # THESE ARE THE ANOMALIES WE ARE LOOKING FOR
# printing anomalies indices 
anomaly_indices_1.shape, anomaly_indices_1
```

## Visualizating normal and abnormal encoded values

```python
encoded_data = autoencoder.encoder(norm_data).numpy()
abnormal = encoded_data[np.where(anomalies_found_1)]
normal = encoded_data[np.where(np.logical_not(anomalies_found_1))]
```

```python
plt.plot(abnormal[:,0],abnormal[:,1], 'x', color='red', label="abnormal")
plt.plot(normal[:,0],normal[:,1], 'o', color='black', label="normal")
plt.grid("on")
plt.legend(loc="upper left")
plt.xlabel("Encoded data component 1")
plt.ylabel("Encoded data component 2")
```

## Solution #2
- We calculate the mean and standard deviation of the reconstruction losses. 
- The losses that are X (arbitray number) standard deviation greater than the mean are considered outliers. Losses less than the mean are closer to zero, therefore are not considered anomalies.

```python
_mean = tf.math.reduce_mean(reconstruction_losses).numpy()
_std = tf.math.reduce_std(reconstruction_losses).numpy()
_mean, _std
```

```python
_dist = reconstruction_losses - _mean
_dist
```

```python
X = 2
anomalies_found_2 = _dist > X * _std
anomaly_indices_2 = np.squeeze(np.where(anomalies_found_2))
anomalies_2 = raw_data[anomaly_indices_2] # THESE ARE THE ANOMALIES WE ARE LOOKING FOR
# printing anomalies indices 
anomaly_indices_2.shape, anomaly_indices_2
```

## Visualizating normal and abnormal encoded values

```python
encoded_data = autoencoder.encoder(norm_data).numpy()
abnormal = encoded_data[np.where(anomalies_found_2)]
normal = encoded_data[np.where(np.logical_not(anomalies_found_2))]
```

```python
plt.plot(abnormal[:,0],abnormal[:,1], 'x', color='red', label="abnormal")
plt.plot(normal[:,0],normal[:,1], 'o', color='black', label="normal")
plt.grid("on")
plt.legend(loc="upper left")
plt.xlabel("Encoded data component 1")
plt.ylabel("Encoded data component 2")
```
